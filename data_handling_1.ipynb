```python
# Columns with missing values 
car_sales.isna().sum()
```

```python
 #Dropping rows with missing columns 
 car_sales = car_sales.dropna(subset=["sellingprice", "saledate", "vin", "mmr"])
```

I dropped these items because they are very critical for the model i want to build, hence it will be misleading for me to make an assumption for them

```python
car_sales.isna().sum()
```

```python
# Filling missing columns 

# Categorical columns
cat_cols = ["make", "model", "trim", "body", "transmission", "color", "interior"]

car_sales[cat_cols] = car_sales[cat_cols].fillna("missing")
```

All non-numerical columns have been filled with "missing"

```python
car_sales.isna().sum()
```

```python
# Filling missing values in the odometer column with the median

car_sales["odometer"] = car_sales["odometer"]. fillna(car_sales["odometer"].median())
```

```python
# Fill missing values in condition with median

car_sales["condition"] = car_sales["condition"]. fillna(
    car_sales["condition"].median())
```

```python
car_sales.isna().sum()
```

```python
car_sales.shape
```

```python
# Group data by a category col

car_sales.groupby("make")["sellingprice"].agg(["mean", "sum", "count"])
```

```python
# Pipeline with at least 4 chained methods

#load data
car_sales_2 = pd.read_csv("data/car_prices.csv")

#Chaining
car_sales_2 = (
    car_sales_2
    .dropna(subset=["sellingprice"])
    .assign(condition=lambda df: df["condition"].fillna(df["condition"].median()))
    .query("odometer < 50000")
    .reset_index()
)
```
